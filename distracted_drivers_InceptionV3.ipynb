# 机器学习进阶毕业项目—驾驶员状态检测—InceptionV3—fine tune

### 一、数据预处理

#### 1、根据司机ID划分训练集和验证集，将P016,P021, P064两位司机，共2315个数据作为验证集，其余25位司机，共20109个数据作为训练集。将验证集从原始文件里移动到验证集相应的分类文件里。

```python
import os
import pandas as pd
import numpy as np
import shutil

valid_subjects = ['p016','p021','P064']

if not os.path.exists("InceptionV3/valid"):
    os.mkdir("InceptionV3/valid")
    for i in range(10):
        os.mkdir("InceptionV3/valid/c%d"%i)

df = pd.read_csv(os.path.join('input', "driver_imgs_list.csv"))

for valid_subject in valid_subjects:
    df_valid = df[(df["subject"]==valid_subject)]
    for index, row in df_valid.iterrows():
        subpath = row["classname"] + "/" + row["img"]
        if os.path.exists(os.path.join('InceptionV3',"train",subpath)):
            shutil.move(os.path.join('InceptionV3',"train",subpath), os.path.join('InceptionV3',"valid",subpath),)
        else:
            print("cannot move {} : {}".format(row["subject"],subpath))
```

### 二、加载数据集

```python
from keras.models import *
from keras.layers import *
from keras.applications import *
from keras.preprocessing.image import *
from keras.applications.inception_v3 import InceptionV3,preprocess_input
np.random.seed(2018)
batch_size=32

gen_train = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    shear_range=0.1,
    zoom_range=0.1,
)

train_generator = gen_train.flow_from_directory(os.path.join('InceptionV3','train'), target_size=(299,299), shuffle=True, batch_size=batch_size,class_mode="categorical")
valid_generator = gen_train.flow_from_directory(os.path.join('InceptionV3','valid'), target_size=(299,299), shuffle=True, batch_size=batch_size,class_mode="categorical")
```

### 三、构建模型

```python
from keras.models import *
from keras.layers import *
from keras.optimizers import *
np.random.seed(2018)

image_size=(299,299)
input_tensor = Input((*image_size, 3))
x = input_tensor
x = Lambda(inception_v3.preprocess_input)(x)

base_model = InceptionV3(input_tensor = x, weights='imagenet',include_top=False)

x = base_model.output
x = GlobalAveragePooling2D()(x) 
x = Dropout(0.5)(x)
predictions = Dense(10,activation='softmax')(x)
model = Model(inputs=base_model.input,outputs=predictions)
```

```python
def fine_tune_a(model,base_model):
    fine_tune_layer_a = 168 
    for layer in base_model.layers[:fine_tune_layer_a-1]:
        layer.trainable = False
    for layer in base_model.layers[fine_tune_layer_a:]:
        layer.trainable = True
    model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
def fine_tune_b(model,base_model):
    fine_tune_layer_b = 168 
    for layer in base_model.layers[:fine_tune_layer_b-1]:
        layer.trainable = False
    for layer in base_model.layers[fine_tune_layer_b:]:
        layer.trainable = True
    model.compile(optimizer=RMSprop(lr=2e-5),loss='categorical_crossentropy',metrics=['accuracy'])
```

### 四、训练模型

#### 1、训练结果可视化，查看模型训练过程中的acc+loss+val_acc+val_loss

```python
import matplotlib.pyplot as plt
%matplotlib inline
def training_vis(hist):
    loss = hist.history['loss']
    val_loss = hist.history['val_loss']
    acc = hist.history['acc']
    val_acc = hist.history['val_acc']

    # make a figure
    fig = plt.figure(figsize=(8,4))
    # subplot loss
    ax1 = fig.add_subplot(121)
    ax1.plot(loss,label='train_loss')
    ax1.plot(val_loss,label='val_loss')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss')
    ax1.set_title('Loss on Training and Validation Data')
    ax1.legend()
    # subplot acc
    ax2 = fig.add_subplot(122)
    ax2.plot(acc,label='train_acc')
    ax2.plot(val_acc,label='val_acc')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Accuracy')
    ax2.set_title('Accuracy  on Training and Validation Data')
    ax2.legend()
    plt.tight_layout()
```

#### 2、fine_tune_a，先从第130层开始。

```python
fine_tune_a(model,base_model)
hist2 = model.fit_generator(generator=train_generator,epochs=1,validation_data=valid_generator)
training_vis(hist2)
```

```python
model.save("weights/InceptionV3-finetune_168_a.h5")
print('model saved')
```

#### 3、fine_tune_b，调小学习率，继续训练。

```python
fine_tune_b(model,base_model)
hist3 = model.fit_generator(generator=train_generator,epochs=2,validation_data=valid_generator)
training_vis(hist3)
```

```python
model.save("weights/InceptionV3-finetune_168_b.h5")
print('model saved')
```

```python
model.compile(optimizer=RMSprop(lr=1e-5),loss='categorical_crossentropy',metrics=['accuracy'])
model.fit_generator(generator=train_generator,epochs=1,validation_data=valid_generator)
```

```python
model.save("weights/InceptionV3-finetune_168_c.h5")
print('model saved')
```

### 六、预测测试集

#### 1、创建提交文件格式。

```python
import datetime
def create_submission(predictions, test_id, info):
    result1 = pd.DataFrame(predictions, columns=['c0', 'c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7', 'c8', 'c9'])
    result1.loc[:, 'img'] = pd.Series(test_id, index=result1.index)
    now = datetime.datetime.now()
    if not os.path.isdir('subm'):
        os.mkdir('subm')
    suffix = info + '_' + str(now.strftime("%Y-%m-%d-%H-%M"))
    sub_file = os.path.join('subm', 'submission_' + suffix + '.csv')
    result1.to_csv(sub_file, index=False)
```

#### 2、获取测试集ID

```python
import glob
def get_test_id():
    X_test_id = []
    print('get test ID')

    path = os.path.join( 'input', 'test','test', '*.jpg')
    files = sorted(glob.glob(path))

    for fl in files:
        flbase = os.path.basename(fl)
        X_test_id.append(flbase)

    return X_test_id
```
